       4. CHAS: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
       5. NOX: nitric oxides concentration (parts per 10 million)
       6. RM: average number of rooms per dwelling
       7. AGE: proportion of owner-occupied units built prior to 1940
       8. DIS: weighted distances to five Boston employment centres
       9. RAD: index of accessibility to radial highways
       10. TAX: full-value property-tax rate per  10,000
       11. PTRATIO: pupil-teacher ratio by town
       12. B: 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
       13. LSTAT: % lower status of the population
       14. MEDV: Median value of owner-occupied homes in 1000's Dolar

### Import File

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy

# Read the csv file
housing_df = pd.read_csv("housing data.csv")
```

```python
# Displaying the first rows, we detect that the column name is missing
housing_df.head()
```

```python
# Now, let's first define the name of the columns
column_names=["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT","MEDV"]

# Read file and assing column names
housing_df = pd.read_csv("housing data.csv", names=column_names)
housing_df.head().T
```

```python
# Split the attached data set into train and test datasets. In this exercise we do not need to split data into 4 sets as
# X_train, X_test, y_train, y_test. The main idea is to evaluate both datasets train and test. We are not going to find 
# any Machine Learning Model to predict MEDV 

import sklearn
from sklearn.model_selection import train_test_split

train, test = train_test_split(housing_df, test_size=0.2, random_state=422)
print('train shape is {}'.format(train.shape))
print('test shape is {}'.format(test.shape))
```

### Review **train** dataset

```python
# Take a picture of the statistical data of dataframe
train.describe().T
```

From these columns we can detect that ZN (proportion of residential land zoned for lots over 25,000 sq.ft.) has 0 for 25% and 50% percentiles, and CHAS has only 2 possible values, which are 1 and 0 (categorical variable).

```python
# Now, let's show the data types and non-null values
train.info()
```

There are zero non-null values

```python
 # Check for zeros in all columns 
train[train == 0].count()    
```

As we saw before ZN and CHAS have many zero values. For ZN (proportion of residential land zoned for lots over 25,000 sq.ft.) many suburbs have no residential land zoned. As these must be considered right values, we do not have to drop the rows associated.
For CHAS zeros, we have seen before that this is a categorical value.

```python
# Visualize each variable's boxplot in order to check the outliers
from scipy import stats

fig, axs = plt.subplots(ncols=7, nrows=2, figsize=(15, 8))
index = 0
axs = axs.flatten()
for k,v in train.items():
    sns.boxplot(y=k, data=train, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

So, we can see that columns CRIM, ZN, RM, B, and MEDV have outliers.

```python
# Lets turn off the warnings due to the use of distplot function.
import warnings
warnings.filterwarnings("ignore")

# The histograms will show us the data distribution
fig, axs = plt.subplots(ncols=7, nrows=2, figsize=(20, 8))
index = 0
axs = axs.flatten()
for k,v in train.items():
    sns.distplot(v, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

On these histograms its shown that some columns ares skewed, as CRIM, ZN, CHAR(because is categorical) anc B. The rest of columns have some kind of normal distribution.

```python
# Finally let's take a look at the correlation
corr = train.corr()
plt.figure(figsize=(15,10))
plt.title("Correlation graph train", fontsize=25)
sns.heatmap(corr, cbar=True, annot=True, square = True, annot_kws={"size": 10})
plt.show()
```

Here we can see some good correlation between some variables and which ones could be good predictors for MEDV

## Exercise #2

Apply some transformation process (standardize numerical data, create dummy columns, polynomials...).

In this case, I'm going to use Normalization preprocessing because is a good technique to use when the distribution is not Gaussian (a bell curve). Standardization is not a good idea, due that the data has is not Gaussian (bell curve) distribution. 

```python
# Normalize preprocessing 
import sklearn
from sklearn.preprocessing import Normalizer

# Standardize features of both datasets
normalizer = Normalizer()
train_norm = normalizer.fit_transform(train)
# Generate data frame to plot new variables
train_norm_df = pd.DataFrame(data=train_norm, columns=train.columns)  
test_norm = normalizer.transform(test)
```

```python
# The histograms will show us the data distribution of normalized train dataset
fig, axs = plt.subplots(ncols=7, nrows=2, figsize=(20, 8))
index = 0
axs = axs.flatten()
for k,v in train_norm_df.items():
    sns.distplot(v, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```
